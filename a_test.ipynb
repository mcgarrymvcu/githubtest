```python
# Step 2: Import Necessary Libraries
from transformers import pipeline
```

```python
# Step 3: Load the Pre-trained Model for Sentiment Analysis
model_name = "distilbert-base-uncased"
nlp_pipeline = pipeline("sentiment-analysis", model=model_name)
```

```python
# Step 4: Provide Input Text and Get Predictions
text = "I am excited to explore artificial intelligence!"
result = nlp_pipeline(text)
```

```python
# Step 5: Print the Output
print("Input Text:", text)
print("Model Output:", result)
```

```python
# Step 6: Test with Another Example
another_text = "I'm not sure if I like AI; it seems complex."
another_result = nlp_pipeline(another_text)
```

```python
print("Input Text:", another_text)
print("Model Output:", another_result)
```

Model Selection:

DistilBERT, the model that was chosen, is a condensed form of BERT (Bidirectional Encoder Representations from Transformers). It was created with the goal of being 60% faster and 40% smaller while maintaining 97% of BERT's language understanding abilities. Because of these characteristics, it is especially well-suited for uses like sentiment analysis that demand great efficiency.
Sentiment analysis, which is frequently employed in domains such as marketing, content moderation, and customer feedback, entails classifying text into positive, negative, or neutral attitudes. Even in settings with limited resources, sentiment analysis may be completed fast and precisely thanks to DistilBERT's lightweight architecture.
